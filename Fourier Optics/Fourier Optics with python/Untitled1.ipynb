```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(k,x):
    return np.exp(-1j*k*x)
```

```python
k_list = np.arange(-40,40,0.01)
```

```python
def mul(k):
    d = 1
    return np.cos(d*2*k)
```

```python
result = 0
x = np.arange(-10,10,0.01)
for k in k_list:
    result += mul(k)*fun(k,x)
```

```python
plt.plot(x,np.real(result))
```
