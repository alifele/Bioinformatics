```python
import numpy as np
%matplotlib notebook
```

```python
import matplotlib.pyplot as plt
```

```python
th = np.arange(-np.pi/2,np.pi/2,0.01)
x = np.arange(-10,10,0.01)
```

```python
def fun(th):
    k = 20
    return np.cos(th)*np.exp(1j*k*np.sin(th)*x)
```

```python
result = 0
for i in th:
    result+=fun(i)
```

```python
plt.plot(np.abs(np.real(result)))
```

```python
patt = np.ones((result.shape[0],result.shape[0]))
```

```python
patt *= np.abs(np.real(result))
```

```python
plt.imshow(patt[900:1100,900:1100])
```

```python
patt = patt.T* np.abs(np.real(result))
patt = patt.T
```

```python
plt.imshow(patt[900:1100,900:1100])
```
